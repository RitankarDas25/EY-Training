```python
import pandas as pd
import numpy as np
```

```python
data = { 
"Month": ["Jan", "Feb", "Mar", "Jan", "Feb", "Mar", "Jan", "Feb", "Mar"], "Product": ["Mobile", "Mobile", "Mobile", "Laptop", "Laptop", "Laptop", "TV", "TV", "TV"], 
"Units_Sold": [200, 220, 210, 120, 135, 150, 80, 95, 100], 
"Revenue": [300000, 330000, 315000, 600000, 675000, 750000, 240000, 285000, 300000] 
} 
df = pd.DataFrame(data) 
df 
```

```python
total_rev_product=df.groupby("Product")["Revenue"].sum()
total_rev_product
```

```python
avg_units_product=df.groupby("Product")["Units_Sold"].mean()
avg_units_product
```

```python
import matplotlib.pyplot as plt

month_order = ['Jan', 'Feb', 'Mar'] # Ensure months are ordered

products = df['Product'].unique() # get unique products
for product in products:
    product_data = df[df['Product'] == product]
    revenues = [product_data[product_data['Month'] == m]['Revenue'].values[0] for m in month_order]
    plt.plot(month_order, revenues, marker='o', label=product)

# Add titles and labels
plt.title("Monthly Revenue Trend by Product (Jan–Mar)")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.legend(title="Product")
plt.show()
```

```python


total_rev_product.plot(kind="bar",color="skyblue",edgecolor="red")

plt.title("Total Revenue per product")
plt.xlabel("Products")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
total_rev_product.plot(
    kind="pie",
    autopct="%1.1f%%",       # Show percentage
    startangle=140,          # Rotate start angle
    colors=plt.cm.Paired.colors       # Optional: use a color palette
)
plt.title("Revenue Contribution by Product")
plt.ylabel("")  # Remove y-label to keep it clean
plt.show()
```
